```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_csv('Crop_recommendation.csv')
```

```python
cleanup_nums={"label":{'rice':0, 'maize':1, 'chickpea':2, 'kidneybeans':3, 'pigeonpeas':4,
       'mothbeans':5, 'mungbean':6, 'blackgram':7, 'lentil':8, 'pomegranate':9,
       'banana':10, 'mango':11, 'grapes':12, 'watermelon':13, 'muskmelon':14, 'apple':15,
       'orange':16, 'papaya':17, 'coconut':18, 'cotton':19, 'jute':20, 'coffee':21}}
data.replace(cleanup_nums,inplace=True)
data.head(10)
```

```python
X=data.iloc[:,0:7].values
y=data.iloc[:,7:].values
y = y.ravel()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
log= RandomForestClassifier()
```

```python
log.fit(X_train,y_train)
pred=log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
a=accuracy_score(y_test,pred)
print("the accuracy of the model: ",a*100)
```

```python
ran=[20,82,6.5,202,5,5.5,98]
ran_arr=np.array(ran)
ran_num=ran_arr.reshape(1,-1)
pred_sin_row=log.predict(ran_num)
print(pred_sin_row)
```
